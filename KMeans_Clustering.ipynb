### K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
#Using the elbow method to find the ideal number of clusters
X = df.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
wcss =[]
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter =300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
from sklearn.preprocessing import MinMaxScaler
df1 = df.drop(columns='Private')
scaler = MinMaxScaler()
scaled_dataset = scaler.fit_transform(df1)
print(type(scaled_dataset))
scaled_dataset
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Cluster=4
# kmeans = KMeans(n_clusters=4,verbose=0,tol=1e-3,max_iter=300,n_init=20)
# kmeans.fit(df.drop('Private',axis=1))

kmeans=KMeans(n_clusters= 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
#Cluster Centers
clus_cent=kmeans.cluster_centers_
clus_cent
```

```python
#Visualising the clusters for k=4
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1],s = 50, c='red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1],s = 50, c='blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1],s = 50, c='green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3,1],s = 50, c='cyan', label = 'Cluster 4')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 200, c = 'yellow', label = 'Centroids')

plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
