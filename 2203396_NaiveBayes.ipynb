```python
#Importing Libraries
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('/pima-indians-diabetes.csv')
```

```python
df = dataset
dataset
```

```python
dataset.describe()
```

```python
df.rename(columns={'1': 'result', '50': 'age', '0.627': 'sugarcount', '33.6': 'haemoglobin', '72': 'blood-oxygen'}, inplace=True)
```

```python
mean_value=df[''].mean()
df['0'].fillna(value=mean_value, inplace=True)
```

```python
features = ['age', 'sugarcount', 'haemoglobin', 'blood-oxygen']
target = 'result'
```

```python
numerical_columns = dataset.select_dtypes(include=[np.number])

correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
columns_to_exclude = ['6', '148', '35']

x = dataset.drop(columns=columns_to_exclude, axis=1)
y = dataset['result']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
for feature in features:
    plt.figure(figsize=(10, 6))
    if x_train[feature].dtype == 'O':
        sns.countplot(data=dataset, x=feature, hue=target)
        plt.title(f'Count Plot of {feature} by {target}')
    else:
        sns.displot(data=dataset, x=feature, hue=target, multiple='stack', kde=True)
        plt.title(f'Distribution Plot of {feature} by {target}')
    plt.xlabel(feature)
    plt.ylabel('result')
    plt.legend()
    plt.show()
```

```python
dataset.dropna(subset=features + [target], inplace=True)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')

x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
classifier = GaussianNB()
classifier.fit(x_train_scaled, y_train)
```

```python
y_test_pred = classifier.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_test_pred)
print(classification_report(y_test, y_test_pred))
```

```python
accuracy = accuracy_score(y_test, y_test_pred)
print(f'Accuracy: {accuracy}')
```
